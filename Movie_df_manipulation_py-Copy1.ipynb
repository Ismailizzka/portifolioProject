```python
# First let's import the packages we will use in this project
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# releod the dataset
df = pd.read_csv(r'C:\Users\ismai\Downloads\movies.csv')
```

```python
# let's check on thr data 

df.head()
```

```python
# check the data types of the columns

df.dtypes
```

```python
# change the data type of the budget into int

df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(np.int64)
```

```python
df['budget'].dtype
```

```python
# change the data type of the gross into int

df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(np.int64)
```

```python
df['gross'].dtype
```

```python
df.dtypes
```

```python
#exrtacting the date of release without the place of release, from the released column

df['dateOfRelease'] = df['released'].str.split('(').str[0]
```

```python
df['dateOfRelease'].str.strip()
```

```python
df.head()
```

```python
# get the year of release from the newly made date of release column because there is some irrigularities with the year column 

df['correctYear'] = df['dateOfRelease'].astype(str).str[-5:]
```

```python
df.head()
```

```python
# sort the data by gross 
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
# looking for a correlletion between gross and budget
```

```python
# Scatter plot with budget vs gross 

plt.scatter(x=df['budget'], y=df['gross'])

plt.xlim(0)

plt.ylim(0)

plt.title('Budget VS Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnigs')

plt.show()
```

```python
# Plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "black"}, line_kws={"color":"purple"})
```

```python
# let's look at the correlation 
```

```python
df.corr(method='pearson') #pearson, kendall, spearman
```

```python
# let's create correlation heatmap

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')
```

```python
#let's check the non_numeric Movie feature correlation 

df_numerized = df.copy()

for col in df_numerized.columns:
    if(df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes

        
df_numerized.head()
```

```python
df.head()
```

```python
# let's create correlation heatmap

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for nonNumeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')
```
